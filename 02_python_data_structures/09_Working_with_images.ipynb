# Working with images
For analysing image data, we need to open them, apply filters to them, segment objects in the image and do measurements. Here we concentrate on visualizing images.

See also
* [Lecture notes on scikit image by Emmanuelle Gouillart](https://scipy-lectures.org/packages/scikit-image/index.html)

## Opening images
Most images read using the [imread](https://scikit-image.org/docs/dev/api/skimage.io.html#skimage.io.imread) function. In case your image doesn't you should consult the documentation of the given file format.

```python
from skimage.io import imread

image = imread("../data/blobs.tif")
```

Images are just matrices of intensities. However, showing them as such is not convenient.

```python
image
```

Images (numpy-arrays) have properties such as dimensionalty, also called `shape`:

```python
image.shape
```

## Visualizing images
We can visualize images using [imshow in scikit image](https://scikit-image.org/docs/dev/api/skimage.io.html#skimage.io.imshow).

```python
from skimage.io import imshow

imshow(image)
```

## Color maps
We can also change the [color map](https://matplotlib.org/stable/tutorials/colors/colormaps.html) (a.k.a. lookup tables) for the visualization.

```python
imshow(image, cmap="Greens_r")
```

```python
imshow(image, cmap="Greens")
```

```python
imshow(image, cmap="gray")
```

```python
imshow(image, cmap="jet")
```
