# How to save results to a csv-file on disc

Creating tables

```python
import numpy as np

# create an empty array: a table is just an array of rows
table = []

for i in range(0, 10):
    # execute analys
    result = i * i
    
    # collect input parameters and result in a new table row
    row = [i, result]
    
    # append a row in the table:
    table = table + [row]

# check if our collection of numbers worked
print(table)
```

```python
# convert the table to a numpy array to make it easier to read
np_table = np.asarray(table)

print(np_table)
```

```python
# save the table to disc. We specify the delimiter to make a COMMA separated file (CSV)

np.savetxt("results.csv", np_table, delimiter=',')
```

# How to read a csv-file from disc

```python
table2 = np.loadtxt("Measurements_ImageJ.csv", delimiter=',')
print(table2)
```

loadtxt has an issue with the first line of our csv file. It contains strings: header of the table.

```python
table2 = np.loadtxt("Measurements_ImageJ.csv", delimiter=',', skiprows=1)
print(table2)
```

```python
# go through the table row by row with a for-loop
for row in table2:
    print(row)
    print("")
```

```python
# go through the table row by row and read the aspect ratio
for row in table2:
    
    aspect_ratio = row[4] # we know AR is in the 5th column
    
    print(aspect_ratio)
```

```python
np.savetxt("Measurements_Python.csv", table2, header=",Area,Mean,Circ.,AR,Round,Solidity", delimiter=',')
```
